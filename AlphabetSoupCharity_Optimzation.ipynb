```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

# Import checkpoint dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint

#  Import and read the charity_data.csv
import pandas as pd 
application_df = pd.read_csv('charity_data.csv')
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=["EIN"], inplace=True)
application_df.head()
```

```python
# Determine the number of unique values in each column.
appli_cat = application_df.dtypes.index.tolist()
application_df[appli_cat].nunique()
```

```python
# Visualize the value counts of CLASSIFICATION
cla = application_df.CLASSIFICATION.value_counts()
cla.plot.density()
```

```python
# Determine which values to replace if counts are less than ..?
replace_class = list(cla[cla < 300].index)

# Replace in dataframe
for cls in replace_class:
    application_df.CLASSIFICATION = application_df.CLASSIFICATION.replace(cls,"Other")
    
# Check to make sure binning was successful
application_df.CLASSIFICATION.value_counts()
```

```python
# Look at NAME value counts for binning
application_df.NAME.value_counts()
```

```python
# Visualize the value counts of NAME
name = application_df.NAME.value_counts()
name.plot.density()
```

```python
# Determine which values to replace if counts are less than ...?
replace_name = list(name[name < 50].index)

# Replace in dataframe
for app in replace_name:
    application_df.NAME = application_df.NAME.replace(app,"Other")
    
# Check to make sure binning was successful
application_df.NAME.value_counts()
```

```python
# Look at ASK_AMT value counts for binning
application_df.ASK_AMT.value_counts()
```

```python
# Visualize the value counts of ASK_AMT
askamt = application_df.ASK_AMT.value_counts()
askamt.plot.density()
```

```python
# Determine which values to replace if counts are less than ...?
replace_askamt = list(askamt[askamt < 3].index)

# Replace in dataframe
for app in replace_askamt:
    application_df.ASK_AMT = application_df.ASK_AMT.replace(app,"Other")
    
# Check to make sure binning was successful
application_df.ASK_AMT.value_counts()
```

```python
application_df.dtypes
```

```python
# Generate our categorical variable lists
application_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_df[application_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(application_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
application_df = application_df.merge(encode_df, left_index=True,right_index=True)
application_df = application_df.drop(columns = application_cat)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL'].values
X = application_df.drop(['IS_SUCCESSFUL'], 1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Optimization Attempt 

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 180
hidden_nodes_layer2 = 60
hidden_nodes_layer3 = 40

# nn stores entire architecture of neural network model
nn = tf.keras.models.Sequential()

nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer1, input_dim = number_input_features,
                         activation = 'relu')
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer2, activation = 'relu')
)

# Third hidden layer
#nn.add(tf.keras.layers.Dense(units = hidden_nodes_layer3, activation = 'relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Check structure of model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Import checkpoint dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint

# Define checkpoint path and filenames
os.makedirs('checkpoints_optimized3/', exist_ok = True)
checkpoint_path = 'checkpoints_optimized3/weights.{epoch:02d}.hdf5'

# Create a callback that saves the model's weights every 5 epochs
cp_callback = ModelCheckpoint(
    filepath = checkpoint_path,
    verbose = 1,
    save_weights_only = True,
    save_freq = 5
)
```

```python
# Train the model
fit_model = nn.fit(
    X_train_scaled, 
    y_train, epochs = 100, 
    callbacks = [cp_callback]
)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export model to HDF5 file
nn.save('AlphabetSoupCharity_Optimization.h5')
```
